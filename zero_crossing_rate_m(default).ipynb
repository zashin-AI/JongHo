```python
import librosa
y, sr = librosa.load('C:/nmb/nmb_data/testvoice_M2.wav')
print(y)
print(y.shape)
print(len(y))
a = librosa.feature.zero_crossing_rate(y)
print(a.shape)
```

```python
zero_crossings = librosa.zero_crossings(y, pad=False)

print(zero_crossings)
print(sum(zero_crossings))
```

```python
import matplotlib.pyplot as plt
n0 = 9000
n1 = 9040

plt.figure(figsize=(16, 6))
plt.title('zero_crossings_m')
plt.plot(y[n0:n1])   # zero_crossings 
plt.grid()

plt.figure(figsize=(16, 6))
plt.title('zero_crossing_rate_m')   
plt.plot(a[0])       # zero_crossing_rate -> 음의 높낮이를 확인할수 있다.   
plt.grid()

plt.show()
```
